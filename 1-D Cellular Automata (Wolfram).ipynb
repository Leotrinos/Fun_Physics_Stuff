```python
import numpy as np
import matplotlib.pyplot as pt
%matplotlib inline
```

<font size = 5, font color = orange> Define the RULE

```python
###### White=0, Black=1
a=0
b=0
c=0
d=1
e=1
f=0
g=1
h=0
```

```python
########## OR using the rule number: 
x=30
a=format(x,'08b')[0]
b=format(x,'08b')[1]
c=format(x,'08b')[2]
d=format(x,'08b')[3]
e=format(x,'08b')[4]
f=format(x,'08b')[5]
g=format(x,'08b')[6]
h=format(x,'08b')[7]
```

<font size = 5, font color = orange> Execute

```python
def Cellular(a,b,c,d,e,f,g,h):
    
    n=100 # n*n plot, change the size! 
    
    ini=np.zeros((n,n),dtype=int)
    ini[0,n//2]=1
    ini[0,n//2+1]=1
    ini[0,n//2-1]=1

    for i in range(0,n-1):  
        for j in range(0,n):
            if 0<j<n-1: 
                if   ini[i,j-1]==1 and ini[i,j]==1 and ini[i,j+1]==1:
                    ini[i+1,j]=a
                elif ini[i,j-1]==1 and ini[i,j]==1 and ini[i,j+1]==0:
                    ini[i+1,j]=b
                elif ini[i,j-1]==1 and ini[i,j]==0 and ini[i,j+1]==1:
                    ini[i+1,j]=c
                elif ini[i,j-1]==1 and ini[i,j]==0 and ini[i,j+1]==0:
                    ini[i+1,j]=d
                elif ini[i,j-1]==0 and ini[i,j]==1 and ini[i,j+1]==1:
                    ini[i+1,j]=e
                elif ini[i,j-1]==0 and ini[i,j]==1 and ini[i,j+1]==0:
                    ini[i+1,j]=f
                elif ini[i,j-1]==0 and ini[i,j]==0 and ini[i,j+1]==1:
                    ini[i+1,j]=g
                elif ini[i,j-1]==0 and ini[i,j]==0 and ini[i,j+1]==0:
                    ini[i+1,j]=h
    return ini

pt.figure(figsize=(15,15))
pt.imshow(Cellular(a,b,c,d,e,f,g,h),cmap='hot',interpolation='None')
pt.title('Rule '+str(int(str(a)+str(b)+str(c)+str(d)+str(e)+str(f)+str(g)+str(h),2)))
# pt.savefig('/Users/Pacifist/Desktop/sth.eps')
pt.show()
```
